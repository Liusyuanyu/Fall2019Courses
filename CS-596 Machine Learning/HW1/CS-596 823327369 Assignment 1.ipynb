```python
import numpy as np
import matplotlib.pyplot as plt
```

## Question 1

```python
#Straight Line y =θ0+θ1*x where θ0=30,θ1=0.5 
point_num= 100
x_val = [xx for xx in range(point_num)]
y_val = list(30+ 0.5*(xx) for xx in range(point_num))
line, = plt.plot(x_val, y_val, lw=2)
plt.title('1-1, Straignt Line')
plt.annotate('straight line', xy=(80, 55), xytext=(20, 30),
            arrowprops=dict(facecolor='black', shrink=0.025),
            )
```

```python
# Quadratic function: y=(x-θ_1 )^2+θ_0,where θ_1=25,θ_0=20
point_start= -50
point_end= 100

x_val = [xx for xx in range(point_start,point_end)]
y_val = list( (xx-25)**2 + 20 for xx in range(point_start,point_end))

plt.title('1-2, Quadratic function')
plt.plot(x_val, y_val, lw=2)
plt.ylim(-1500,7000)
plt.annotate('min value', xy=(25, 20), xytext=(-10, -1000),
            arrowprops=dict(facecolor='black', width = 0.5, headwidth=5),
            )
```

```python
# Log function, y=-log⁡(x) and y=-log⁡(1-x) 
point_start= 1
point_end= 100

x_val = [xx/100 for xx in range(point_start,point_end)]
y_val = list( np.log10(xx/100) for xx in range(point_start,point_end))

plt.title('1-3, Log function')
plt.plot(x_val, y_val, lw=2)

y_val = list( -np.log10(1 - xx/100) for xx in range(point_start,point_end))
plt.plot(x_val, y_val, lw=2)

plt.ylim(-2.5,2.5)
plt.annotate('log(x)', xy=(0.4,-0.8), xytext=(0.6, -1.5),
            arrowprops=dict(facecolor='black', width = 0.5, headwidth=8),
            )
plt.annotate('-log(1-x)', xy=(0.5,0.4), xytext=(0.2, 1),
            arrowprops=dict(facecolor='black', width = 0.5, headwidth=8),
            )
```

```python
# Sigmoid function, y=1/(1+e^(-x))
point_start= -10
point_end= 11

x_val = [xx for xx in range(point_start,point_end)]
y_val = list( 1/(1 + np.exp(-xx)) for xx in range(point_start,point_end))

plt.title('1-4, Sigmoid function')
plt.plot(x_val, y_val, lw=2)
plt.ylim(-0.2,1.2)
plt.xlim(-12,12)

## Lables
plt.annotate('approach zero', xy=(-9, 0.05), xytext=(-7.5, 0.2),
            arrowprops=dict(facecolor='black', width = 0.5, headwidth=10),
            )
plt.annotate('approach one', xy=(9, 0.95), xytext=(3, 0.7),
            arrowprops=dict(facecolor='black', width = 0.5, headwidth=10),
            )
```
